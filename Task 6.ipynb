### Importing necessary libraries

```python
import numpy as n
import pandas as p
import seaborn as s
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore") 

from sklearn.tree  import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn import tree
```

### Importing Dataset

```python
data = p.read_csv('Iris.csv')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

### Checking for Duplicated values

```python
data.duplicated()
```

There are no duplicated/redundant rows

### Checking for Null values

```python
data.isnull()
```

There are no null/missing values

### Dropping column 'Id'

```python
df = data.drop(['Id'], axis=1)
```

### Heatmap of Correlation matrix

```python
fig, ax = plt.subplots(figsize=(10,7))
s.heatmap(df.corr(), annot=True)
```

We see a good correlation between PetalLengthCm and PetalWidthCm, 
PetalLengthCm and SepalLengthCm, SepalLengthCm and PetalWidthCm

### Splitting dataset

```python
x = df.drop('Species', axis=1)
y = df["Species"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

x is input, y is output and train size is 80%, test size is 20%

### Training the model

```python
dt = DecisionTreeClassifier(random_state=0)
dt.fit(x_train,y_train)
print('Training is complete')
```

### Testing the model

```python
y_pred = dt.predict(x_test)
print('Testing is complete')
df1 = p.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df1
```

### Evaluating model performance

```python
print('Accuracy:', accuracy_score(y_pred, y_test))
print(classification_report(y_test, y_pred))
print(s.heatmap(confusion_matrix(y_test, y_pred), annot=True))
```

We achieved very high precision, recall and f1 score, which says that our model fit and predictions are very good

### Visualizing the model

```python
feature_names=['SepalLength(Cm)', 'SepalWidth(cm)', 'PetalLength(cm)', 'PetalWidth(cm)']
target_names=['Setosa', 'Versicolor', 'Virginica']
fig=plt.figure(figsize=(20,15))
tree.plot_tree(dt, feature_names=feature_names, class_names=target_names,filled=True)
```
